# ADD NEIGHBORHOOD TO PROPERTY VIA SPATIAL JOIN
NOTE: THIS IS DONE BECAUSE THE ORIGINAL 'NEIGHBORHOODS' COLUMN HAS SEVERAL REGIONS AND NEIGHBORHOODS ASSIGNED TO ONE LISTING (EG. "Fairfax District, Central LA, Beverly Fairfax") MAKING IT UNUSABLE FOR THE MODEL

```python
import pandas as pd
import zipfile
import io, os
import geopandas as gpd
```

```python
notebook_dir = os.getcwd()
pd.set_option('display.max_columns', 500)
```

```python
#IMPORT LISTING DATA AND NEIGHBORHOOD DATA INTO GEODATAFRAMES
df = pd.read_csv("../data/Los_Angeles_Sold_20240829.csv")
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.longitude, df.latitude), crs="EPSG:4326"
)

gdf_neighborhoods = gpd.read_file("../data/neighbourhoods.geojson")
```

```python
#PERFORM SPATIAL JOIN
gdf_spatial_join = gpd.sjoin(gdf, gdf_neighborhoods, how='inner', predicate='within')
```

```python
#DROP UNUSED COLUMNS
gdf_spatial_join = gdf_spatial_join.drop(columns=['index_right', 'neighbourhood_group'])
```

```python
#EXPORT LISTINGS DATA
gdf.to_csv('../data/Los_Angeles_Sold_spatial_join.csv', index=False)
```
